```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer
from datasets import load_dataset

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
model.config.pad_token_id = tokenizer.pad_token_id

# Load dataset
import pandas as pd
from datasets import Dataset

# Load CSV from Google Drive using pandas
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/living_things_dataset.csv")

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(df)


# Combine and tokenize
def tokenize_function(example):
    text = example["prompt"] + " " + example["text"]
    tokens = tokenizer(text, truncation=True, padding="max_length", max_length=512)
    tokens["labels"] = tokens["input_ids"].copy()
    return tokens

tokenized_dataset = dataset.map(tokenize_function)

# Set training arguments
training_args = TrainingArguments(
    output_dir="./gpt2-living-thing",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    logging_steps=10,
    save_steps=30,
    save_total_limit=1,
    fp16=True,             # Mixed precision on GPU
    report_to="none"       # Disable wandb
)

# Initialize trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

# Start training
trainer.train()
```

```python
# Step 1: Install dependencies if not already installed
!pip install transformers datasets

# Step 2: Imports
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer

# Step 3: Load CSV manually from Google Drive
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/living_things_dataset.csv")

# Step 4: Convert to Hugging Face dataset
dataset = Dataset.from_pandas(df)

# Step 5: Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # Important: GPT-2 has no pad token by default

model = AutoModelForCausalLM.from_pretrained("gpt2")
model.config.pad_token_id = tokenizer.pad_token_id

# Step 6: Tokenize dataset
def tokenize_function(example):
    full_input = example["prompt"] + " " + example["text"]
    tokens = tokenizer(
        full_input,
        truncation=True,
        padding="max_length",
        max_length=128
    )
    tokens["labels"] = tokens["input_ids"].copy()
    return tokens

tokenized_dataset = dataset.map(tokenize_function)

# Step 7: Set training arguments
training_args = TrainingArguments(
    output_dir="./gpt2-living-thing",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    logging_steps=10,
    save_steps=30,
    save_total_limit=1,
    fp16=True,  # Enable mixed precision if GPU available
    report_to="none"  # Avoid W&B login issues
)

# Step 8: Initialize trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer
)

# Step 9: Train the model
trainer.train()
```

```python
import torch

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Input prompt
input_text = "Describe a living thing. "

# Tokenize with attention mask and move tensors to same device
inputs = tokenizer(
    input_text,
    return_tensors="pt",
    padding=True,
    truncation=True
)
inputs = {key: val.to(device) for key, val in inputs.items()}

# Generate text
output = model.generate(
    **inputs,
    max_length=50,
    do_sample=True,
    temperature=0.8,
    top_p=0.9,
    pad_token_id=tokenizer.eos_token_id  # set pad token to EOS
)

# Decode and print
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
